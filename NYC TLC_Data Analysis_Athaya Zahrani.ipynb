# **A. *BACKGROUND***

*`New York City Taxi and Limousine Commission (TLC)`* adalah sebuah perusahaan/badan yang dibawahi oleh pemerintahan kota New York di Amerika Serikat yang mencakup industri sewa kendaraan, mengatur, serta memfasilitasi Taksi/Limousin yang bisa digunakan untuk warga kota atau wisatawan di kota New York.

TLC didirikan pada tahun 1971 oleh John Lindsay sebagai walikota yang menjabat pada periode tersebut. Perusahaan ini didirikan dengan tujuan untuk memajukan transportasi antar kota.

Pada saat ini, TLC telah memberi sekitar 170.000 lisensi kepada driver profesional, menggunakan 100.000 lebih kendaraan, dan memiliki kurang lebih 1000 pangkalan sewa kendaraan yang tersebar di setiap sudut kota New York. 

Salah satu ciri khas dari taksi ini adalah warnanya yang mencolok yaitu kuning, ini sudah menjadi ciri khas atau ikon kota New York, dan sering dijumpai di beberapa film yang berlatar di New York, seperti Home Alone, Baby's Day Out, Taxi Driver, dan The Irishman.

## 3. Distribusi Data Kolom Numerikal

```python
numerical = ['trip_distance', 'fare_amount', 'tip_amount',
       'tolls_amount', 'extra', 'mta_tax', 'improvement_surcharge',
       'congestion_surcharge', 'total_amount', 'total_final',
       'trip_duration', 'driving_speed', 'passenger_count']
```

```python
from scipy.stats import normaltest

dist = []

for i in range(len(numerical)):
    feature_name = numerical[i]
    stat, p_value = normaltest(df[feature_name].dropna())
    
    if p_value > 0.05:
        distribution = "Normally Distributed"
    else:
        distribution = "Not Normally Distributed"
    
    dist.append([feature_name, stat, p_value, distribution])

dist_df = pd.DataFrame(dist, columns=['Feature', "D'Agostino-Pearson Statistic", 'P-value', 'Distributed'])
dist_df
```

```python
plt.figure(figsize=(18, 18))

for index, col in enumerate(numerical):
    # Histogram
    plt.subplot(len(numerical), 2, index * 2 + 1)
    sns.histplot(df[col], kde=True, color='#FA7070', bins=50)
    plt.title(f'Histogram of {col}', size=12)
    plt.xlabel(col)
    plt.ylabel('Frequency')

    # Boxplot
    plt.subplot(len(numerical), 2, index * 2 + 2)
    sns.boxplot(x=df[col], color='#FFC700')
    plt.title(f'Boxplot of {col}', size=12)
    plt.xlabel(col)

plt.tight_layout()
plt.show()
```

> Seluruh kolom numerikal tidak berdistribusi normal.

## 1. Kapan saja tanggal di bulan Januari 2023 yang memiliki permintaan tertinggi?

```python
def despine(ax):

    for spine in ['top', 'right']:
        ax.spines[spine].set_visible(False)
    
    return ax
```

```python
# Ambil data tanggal di bulan Januari
df["pickup_date"] = df["lpep_pickup_datetime"].dt.day
grouped_df_date = df.groupby("pickup_date").size().reset_index(name="trip_count")
grouped_df_date.head() 
```

```python
plt.figure(figsize=(12, 6))
plt.plot(grouped_df_date["pickup_date"], grouped_df_date["trip_count"], 'bo--')

plt.xticks(ticks=grouped_df_date["pickup_date"])
plt.xlabel("Tanggal Penjemputan")
plt.ylabel("Jumlah Perjalanan")
plt.title("Permintaan NYC TLC di Bulan Januari 2023")

for x, y in zip(grouped_df_date["pickup_date"], grouped_df_date["trip_count"]):
    plt.annotate(
        text = f'{y:,.0f}',    # teks yang ditampilkan
        xy = (x, y),           # koordinat titik yang dianotasi
        xytext = (0, 10),      # offset teks anotasi dari titik data (dalam pixel)
        textcoords = 'offset points',
        ha = 'center'          # horizontal alignment
    )

plt.show()
```

```python
# Menampilkan 5 tanggal dengan jumlah perjalanan tertinggi
print("5 tanggal dengan jumlah perjalanan tertinggi: ")
grouped_df_date_peak = grouped_df_date.sort_values("trip_count", ascending = False).head(5)
grouped_df_date_peak
```

- Jumlah permintaan NYC TLC yang rendah pada bulan Januari 2023 terjadi pada tanggal 1, 2, 8, 16, dan 29. ## 2. Bagaimana variasi waktu penjemputan (*pick up hour*) yang memiliki permintaan NYC TLC tertinggi?

```python
# Berdasarkan waktu penjemputan
df["pickup_hour"] = df["lpep_pickup_datetime"].dt.hour
grouped_df_hour = df.groupby("pickup_hour").size().reset_index(name="trip_count")
grouped_df_hour.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(grouped_df_hour["pickup_hour"], grouped_df_hour["trip_count"], 'bo--')

plt.xticks(ticks=grouped_df_hour["pickup_hour"])
plt.xlabel("Waktu Penjemputan")
plt.ylabel("Jumlah Perjalanan")
plt.title("Permintaan NYC TLC Berdasarkan Waktu Penjemputan di Bulan Januari 2023")

for x, y in zip(grouped_df_hour["pickup_hour"], grouped_df_hour["trip_count"]):
    plt.annotate(
        text = f'{y:,.0f}',    # teks yang ditampilkan
        xy = (x, y),           # koordinat titik yang dianotasi
        xytext = (0, 10),      # offset teks anotasi dari titik data (dalam pixel)
        textcoords = 'offset points',
        ha = 'center'          # horizontal alignment
    )

plt.show()
```

**Insight**
- Jumlah permintaan NYC TLC mulai naik di rentang pukul 05:00 sampai 09:00 dan pukul 13:00 sampai 15:00. Periode permintaan ini sesuai dengan jam sibuk pagi dan sore untuk berangkat sekolah atau pergi bekerja.
- Jumlah permintaan NYC TLC turun drastis mulai pukul 18:00 hingga larut malam, dilanjut pada pukul 00:00 hingga 05:00 hingga mencapai titik minimum. Hal ini memungkinkan bahwa masyarakat sudah pulang ke rumah.

```python
pivot_table_fare_distance_duration = grouped_df_fare_distance_duration.pivot(index="distances_range", columns="trip_duration_range", values="fare_amount")
pivot_table_fare_distance_duration
```

```python
# Menentukan urutan rentang jarak dan durasi
desired_distance_order = ['>30 miles', '21-30 miles', '11-20 miles', '<10 miles']
desired_duration_order = ['<10 minutes', '11-20 minutes', '21-30 minutes', '31-40 minutes', '41-50 minutes', '>50 minutes']

# Mengubah indeks dan kolom pivot table menjadi kategori terurut
pivot_table_fare_distance_duration.index = pd.CategoricalIndex(pivot_table_fare_distance_duration.index, categories=desired_distance_order, ordered=True)
pivot_table_fare_distance_duration.columns = pd.CategoricalIndex(pivot_table_fare_distance_duration.columns, categories=desired_duration_order, ordered=True)
pivot_table_fare_distance_duration = pivot_table_fare_distance_duration.sort_index(axis=0).sort_index(axis=1)

plt.figure(figsize = (6, 3))
sns.heatmap(pivot_table_fare_distance_duration, cmap = "Blues", annot = True, fmt = ".2f")

plt.xlabel("Rentang Durasi Perjalanan")
plt.xticks(rotation = 45)
plt.ylabel("Rentang Jarak Perjalanan")
plt.title("Rata-rata Fare Amount Berdasarkan Jarak dan Durasi Perjalanan pada Bulan Januari 2023")
plt.show()
```

**Insight**
- Secara umum, berdasarkan *heat map*, meningkatnya jarak sejalan dengan meningkatnya durasi perjalanan, juga diikuti juga oleh meningkatnya tarif rata-rata yang dikeluarkan oleh meteran kendaraan.
